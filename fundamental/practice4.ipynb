```python
import numpy as np
import seaborn as sb
import matplotlib.pyplot as plt
import pandas as pd
```

```python
x = np.arange(0,10,0.01)
```

```python
y1 = x-5
```

```python
print(len(y1))
```

```python
y2 = -x+5
```

```python
def intersection(x,y1,y2):
    intersection = []
    for i in range(len(y1)):
        if abs(y1[i]-y2[i]) < 0.01:
            intersection.append((x[i],y1[i]))
    return intersection
```

```python
intersection_result = intersection(x,y1,y2)
titikSinggung = pd.DataFrame(intersection_result, columns=["X","Y"])
display(titikSinggung)
```
