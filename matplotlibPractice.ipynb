```python
PI = np.pi
```

```python
sudut = np.linspace(0,2*PI,100) #pendistribusian elemen ke array secara merata dari 0,360,sebanyak 100
```

```python
radius = 5
```

```python
x = radius * np.cos(sudut)
```

```python
x
```

```python
y = radius * np.sin(sudut)
```

```python
plt.plot(x,y)
plt.show()
```

```python
x = np.linspace(-20,20,100)
```

```python
y1 = 2*x**2 + 3*x + 5
```

```python
y2 = 2*x + 1
```

```python
df = pd.DataFrame({
    'x' : x,
    'y1' : y1,
    'y2' : y2
})
```

```python
display(df)
```

```python
plt.plot(df['x'],df['y1'],label='y1 = 2x^2+3x+5')
plt.plot(df['x'],df['y2'],label='y2 = 2x + 1')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Equation of Tangent Line')
plt.legend()
plt.grid(True)
plt.show()
```
