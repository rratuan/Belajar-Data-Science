```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
sudut = np.arange(-np.pi,np.pi,0.01)
```

```python
radius = 5
```

```python
x = radius * np.cos(sudut)
```

```python
print(x)
```

```python
y1 = radius * np.sin(sudut)
```

```python
y2 = x+1
```

```python
def intersection(x,y1,y2):
    intersection_points = []
    for i in range(len(y1)):
        if abs(y1[i]-y2[i]) < 0.03:
            intersection_points.append((x[i],y1[i]))
    return intersection_points
```

```python
intersection_result = intersection(x,y1,y2)
intersection = pd.DataFrame(intersection_result,columns=["x","y"])
display(intersection)
```

```python
plt.plot(x,y1)
plt.plot(x,y2)
plt.scatter(intersection['x'],intersection['y'])
plt.show()
```
