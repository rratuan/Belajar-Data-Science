```python
import seaborn as sb
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.arange(0,10,0.01)
```

```python
y1 = np.sin(x) * 10
```

```python
y2 = x
```

```python
def intersection(x,y1,y2):
    intersection = []
    for i in range(len(y1)):
        if abs(y1[i]-y2[i]) < 0.04:
            intersection.append((x[i],y1[i]))
    return intersection
```

```python
intersection_result = intersection(x,y1,y2)
titikSinggung = pd.DataFrame(intersection_result, columns=["X","Y"])
display(titikSinggung)
```

```python
plt.plot(x,y1)
plt.plot(x,y2,color='green')
plt.scatter(titikSinggung['X'],titikSinggung['Y'])
```
